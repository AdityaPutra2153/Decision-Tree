```python
import numpy as np
import pandas as pd
from sklearn import tree
```

# Membaca dataset dari file ke pandas dataframe

```python
FishDataset = pd.read_csv('Fish.csv')
```

```python
FishDataset.head()
```

# Mengubah keals (kolom "Species") dari String ke Unique-Integer

```python
FishDataset["Species"] = pd.factorize(FishDataset.Species)[0]
```

```python
FishDataset.head()
```

```python
print(FishDataset)
```

# Mengubah dataFrame ke Array Numpy

```python
FishDataset = FishDataset.to_numpy()
```

```python
print(FishDataset)
```

# Membagi dataset 
membagi dataset untuk data training yaitu sebanyak 120 baris untuk data training dan untuk data testing sebanyak 30 baris.

```python
dataTraining = np.concatenate((FishDataset[0:60, :], FishDataset[75:135, :]), 
                              axis=0)
dataTesting = np.concatenate((FishDataset[60:75, :], FishDataset[135:150, :]),
                              axis=0)
```

```python
print(dataTraining)
len(dataTraining)
```

```python
print(dataTesting)
len(dataTesting)
```

# Memecah Dataset ke Input dan Label

```python
inputTraining = dataTraining[:, 1:6]
inputTesting = dataTesting[:, 1:6]
labelTraining = dataTraining[:, 0]
labelTesting = dataTesting[:, 0]
```

# Mendefinisikan Decision Tree Classifier

```python
model = tree.DecisionTreeClassifier()
```

# Mentraining Model

```python
model = model.fit(inputTraining, labelTraining)
```

# Memprediksi Input Data Testing

```python
hasilPrediksi = model.predict(inputTesting)
print ("Label Sebenarnya : ", labelTesting)
print("Hasil Prediksi   : ", hasilPrediksi)
```

# Menghitung Akurasi

```python
prediksiBenar = (hasilPrediksi == labelTesting).sum()
prediksiSalah = (hasilPrediksi != labelTesting).sum()
print("Prediksi Benar   : ", prediksiBenar, "data")
print("Prediksi Salah   : ", prediksiSalah, "data")
print("Akurasi          : ", prediksiBenar/(prediksiBenar+prediksiSalah) * 100, "%")
```
